```python
f = open('./data/대파제육볶음.txt','r', encoding='utf8')
txt = f.readlines()
f.close()
```

```python
new_txt = []
for i in txt:
    new_txt.append(i.replace("'",''))
```

```python
a = word.pos(new_txt[0], join=True)
```

```python
N = []
n = []
for i in a:
    if i[-4:] == 'Noun':
        N.append(i)
    else:
        pass
    
for k in range(0,513):
    x = N[k][:-5]
    n.append(x)
n
```
