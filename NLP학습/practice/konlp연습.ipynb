```python
from konlpy.tag import Okt
```

```python
word = Okt()
```

```python
text = '한국어 분석을 시작합니다.'

# 정규화(normalization) 처리
print('정규화 처리 결과 =',word.normalize(text))
print()

# 어구 추출(Phrases Extraction)
print('어구 추출 결과 =',word.phrases(text))
print()

# 형태소 분석
print('형태소 분석 =',word.morphs(text))
print()

# 형태소 분석(Pos Tagger)
print('형태소 분석(Pos Tagger) =',word.pos(text))
print()
print('형태소 분석(Pos Tagger) / =',word.pos(text, join=True))
print()

# 명사(Noun) 추출
print('명사 추출 결과 =',word.nouns(text))
```
