```python
from pydub import AudioSegment

choi_file = AudioSegment.from_mp3("data/choi.mp3")
type(choi_file)

choi_file.export(out_f="data/choi.wav", format="wav")

choi_audio = sr.AudioFile("data/choi.wav")

with choi_audio as source:
    choi_jtbc = recognizer.record(source)

recognizer.recognize_google(audio_data=choi_jtbc, language="ko-KR")
```

```python
from pydub import AudioSegment
import math
song = AudioSegment.from_mp3("성공률 100_ 돼지고기 김치찌개.wav")

sixty_seconds = 240 * 1000

for i in range(int(math.floor(len(song)/240000))):
    slice = song[i*sixty_seconds:sixty_seconds*(i+1)]
    slice.export('sample3/newSong_{}.wav'.format(i), format="wav")
    
# choi_audio = sr.AudioFile('newSong_{}.wav')
import time
start=time.time()
text = []

for i in range(0,15,1):
    choi_audio = sr.AudioFile('sample/newSong_{}.wav'.format(i))
    with choi_audio as source:
            choi_jtbc = recognizer.record(source)
#             recognizer.recognize_google(audio_data=choi_jtbc, language="ko-KR")
            text.append(recognizer.recognize_google(audio_data=choi_jtbc, language="ko-KR"))
print(text)
print(time.time()-start)
```
