```python
import pandas as pd
```

```python
hmh = pd.read_excel('./data/긍정_허무함.xlsx')
sc = pd.read_excel('./data/긍정_상처.xlsx')
dh = pd.read_excel('./data/긍정_당황.xlsx')
happy = pd.read_excel('./data/행복증강.xlsx')
```

```python
aug = pd.concat([hmh, sc, dh, happy], axis=0)
aug = aug.reset_index(drop=True)
aug
```

```python
df = pd.read_excel('./data/긍정일상문.xlsx')
df
```

```python
DF = pd.concat([df,aug], axis=0)
DF = DF.reset_index(drop=True)
DF
```

```python
DF = DF.drop_duplicates(['word'], keep='first')
DF = DF.reset_index(drop=True)
DF
```

```python
DF.to_excel('./data/긍정일상문_증강.xlsx', index=False)
```

```python
dic = pd.read_excel('./data/최종사전_증강(부정).xlsx')
dic
```

```python
fdic = pd.concat([dic,DF], axis=0)
fdic = fdic.reset_index(drop=True)
fdic
```

```python
fdic = fdic.drop_duplicates(['word'], keep='first')
fdic = fdic.reset_index(drop=True)
fdic
```

```python
fdic.to_excel('./data/최종사전_증강(긍부정).xlsx', index=False)
```

```python
fdic = fdic.dropna()
fdic
```
