```python
import pandas as pd
```

```python
hmh = pd.read_excel('./data/허무함.xlsx')
sc = pd.read_excel('./data/상처.xlsx')
dh = pd.read_excel('./data/당황.xlsx')
```

```python
aug = pd.concat([hmh, sc, dh], axis=0)
aug = aug.reset_index(drop=True)
aug
```

```python
df = pd.read_excel('./data/부정일상문.xlsx')
df
```

```python
DF = pd.concat([df,aug], axis=0)
DF = DF.reset_index(drop=True)
DF
```

```python
DF = DF.drop_duplicates(['word'], keep='first')
DF = DF.reset_index(drop=True)
DF
```

```python
DF.to_excel('./data/일상문+증강.xlsx', index=False)
```

```python
dic = pd.read_excel('./data/최종감성사전.xlsx')
dic
```

```python
fdic = pd.concat([dic,DF], axis=0)
fdic = fdic.reset_index(drop=True)
fdic
```

```python
fdic = fdic.drop_duplicates(['word'], keep='first')
fdic = fdic.reset_index(drop=True)
fdic
```

```python
fdic.to_excel('./data/최종사전_증강(부정).xlsx', index=False)
```
