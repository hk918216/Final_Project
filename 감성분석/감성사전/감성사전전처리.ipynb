```python
import pandas as pd
import json
```

```python
with open('./data/SentiWord_info.json', encoding='utf-8-sig', mode='r') as f:
            data = json.load(f)
```

```python
data
```

```python
len(data)
```

```python
word = []
word_root = []
polarity = []

for i in range(len(data)):
    word.append(data[i]['word'])
    word_root.append(data[i]['word_root'])
    polarity.append(data[i]['polarity'])
```

```python
emo_dic = pd.DataFrame({'word' : word, "word_root":word_root, "polarity":polarity})
emo_dic
```

```python
emo_dic.to_csv('./data/감성사전made.csv', index=False)
```
