```python
import pandas as pd
```

## 긍정/부정 전처리

```python
pp = pd.read_excel('./data/매우긍정.xlsx')
pp
```

```python
pn = pd.read_csv('./data/매우부정.csv')
pn
```

### value 확인 및 null 처리

```python
pp['emotion'].unique()
```

```python
ppos = pp.drop(['polarity'], axis =1)
ppos
```

```python
# emotion vlaue 값 확인
pn['emotion'].unique()
```

```python
# null 값 삭제
pneg = pn.dropna()
```

```python
# polarity 컬럼 삭제
pneg = pneg.drop(['polarity'], axis =1)
pneg
```

## 데이터프레임 합치고 값 부여

```python
ppn = pd.concat([ppos, pneg], axis=0)
ppn
```

```python
# index 재설정
ppn = ppn.reset_index(drop=True)
```

```python
# emotion 값 리스트로 만들기
list = []
for i in ppn['emotion']:
    list.append(i)
```

```python
# list 값 확인하며 고유값 부여
emo_num = []

for i in range(0,len(list)):
    if list[i] == '분노' :
        emo_num.append(6)
    elif list[i] == '슬픔':
        emo_num.append(7)
    elif list[i] == '당황':
        emo_num.append(8)
    elif list[i] == '상처':
        emo_num.append(9)
    elif list[i] == '불안':
        emo_num.append(10)
    elif list[i] == '짜증':
        emo_num.append(11)
    elif list[i] == '답답':
        emo_num.append(12)
    elif list[i] == '허무함':
        emo_num.append(13)
    elif list[i] == '조롱':
        emo_num.append(14)
    elif list[i] == '불편':
        emo_num.append(15)
    elif list[i] == '걱정':
        emo_num.append(16)
    elif list[i] == '불만':
        emo_num.append(17)
    elif list[i] == '무서움':
        emo_num.append(18)
    elif list[i] == '편안':
        emo_num.append(0)
    elif list[i] == '기쁨':
        emo_num.append(1)
    elif list[i] == '행복':
        emo_num.append(2)
    elif list[i] == '감사':
        emo_num.append(3)
    elif list[i] == '만족':
        emo_num.append(4)
    elif list[i] == '흥미':
        emo_num.append(5)
    else :
        pass
```

```python
emo_num = pd.DataFrame(emo_num)
emo_num
```

```python
df = pd.concat([ppn, emo_num], axis=1)
```

```python
# 컬럼명 재설정
df.columns=['word','word_root','emotion','emo_num']
```

```python
df
```

```python
# 파일저장
df.to_excel('./data/매우긍정_매우부정.xlsx', index=False)
```
